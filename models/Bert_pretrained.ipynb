### 2.1 Loading in a dataset

In addition to having models, the [the hub](https://huggingface.co/datasets) also has datasets.

```python
import pandas as pd

TRAIN = "/content/drive/MyDrive/LUN_data/raw_data/fulltrain.csv"
TEST = "/content/drive/MyDrive/LUN_data/raw_data/balancedtest.csv"

def prepare_data(pathi):
  df = pd.read_csv(pathi, header=None)
  df.columns = ["reliability", "text"]
  df.reliability = df.reliability.astype(int)
  df.text = df.text.astype(str)
  return df

train_set = prepare_data(TRAIN)
test_set = prepare_data(TEST)

print(train_set.head())
print(test_set.head())
```

```python
from sklearn.model_selection import train_test_split


train_texts, val_texts, train_labels, val_labels = train_test_split(train_set.text.values.tolist(), train_set.reliability.values.tolist(), test_size=.2)
test_texts = test_set.text.values.tolist()
test_labels = test_set.reliability.values.tolist()
```

```python
from transformers import DistilBertTokenizerFast
tokenizer = DistilBertTokenizerFast.from_pretrained('distilbert-base-cased')
```

```python
train_encodings = tokenizer(train_texts, truncation=True, padding=True)
val_encodings = tokenizer(val_texts, truncation=True, padding=True)
test_encodings = tokenizer(test_texts, truncation=True, padding=True)
```

```python
train_encodings[1]
```

```python
import torch

class LUNDataset(torch.utils.data.Dataset):
    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels

    def __getitem__(self, idx):
        item = {key: torch.tensor(val[idx]) for key, val in self.encodings.items()}
        item['labels'] = torch.tensor(self.labels[idx] - 1)
        return item

    def __len__(self):
        return len(self.labels)

train_dataset = LUNDataset(train_encodings, train_labels)
val_dataset = LUNDataset(val_encodings, val_labels)
test_dataset = LUNDataset(test_encodings, test_labels)
```
